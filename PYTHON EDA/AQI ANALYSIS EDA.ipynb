# 2. What is AQI?

**AQI stands for Air Quality Index.**

It is a numerical value that represents the quality of air in a specific location.  
AQI tells us whether the air is safe or harmful to breathe.

A lower AQI value means better air quality.  
A higher AQI value means polluted air and greater health risk.

AQI is usually calculated using major air pollutants such as:

| Pollutant | Full Form | Simple Meaning |
|---|---|---|
| PM2.5 | Particulate Matter 2.5 | Very tiny dust particles that can enter deep into the lungs |
| PM10 | Particulate Matter 10 | Larger dust particles found in polluted air |
| NO₂ | Nitrogen Dioxide | Gas mainly released from vehicles and industries |
| SO₂ | Sulphur Dioxide | Gas released from burning fossil fuels |
| CO | Carbon Monoxide | Harmful gas produced by incomplete burning of fuel |
| O₃ | Ozone | Gas that can irritate lungs when present near the ground |



## Standard AQI Categories

| AQI Range | Air Quality Category | Meaning | Possible Health Impact |
|---:|---|---|---|
| 0 - 50 | Good | Air is clean | Little or no health risk |
| 51 - 100 | Satisfactory / Moderate | Air quality is acceptable | Sensitive people may feel slight discomfort |
| 101 - 200 | Moderately Polluted | Pollution is noticeable | Breathing discomfort for sensitive groups |
| 201 - 300 | Poor | Air is unhealthy | People may experience breathing problems |
| 301 - 400 | Very Poor | Air is highly polluted | Serious respiratory problems may occur |
| 401 - 500 | Severe | Air is dangerous | Serious health risk for everyone |

# 17. Create Pollution Category Column

````python
# This function creates a simple pollution category using pollutant average values.
# Note: This is a simple category for EDA understanding, not an official AQI calculation.
def pollution_category(value):
    if value <= 50:
        return 'Good'
    elif value <= 100:
        return 'Moderate'
    elif value <= 200:
        return 'Poor'
    elif value <= 300:
        return 'Very Poor'
    else:
        return 'Severe'

# Apply the function to pollutant avg column.
aqi_df['pollution_category'] = aqi_df['pollutant avg'].apply(pollution_category)

# Check category counts.
aqi_df['pollution_category'].value_counts()
````

````python
# Plot pollution category count.

category_order = ['Good', 'Moderate', 'Poor', 'Very Poor', 'Severe']
category_count = aqi_df['pollution_category'].value_counts().reindex(category_order, fill_value=0)

plt.figure(figsize=(8,5))

category_count.plot(
    kind='barh',          # Horizontal bar chart
    color='purple',       # Purple bars
    edgecolor='black',    # Black border
    linewidth=1.2         # Border thickness
)

plt.title('Pollution Category Count')
plt.xlabel('Number of Records')
plt.ylabel('Pollution Category')

plt.grid(axis='x', linestyle='--', alpha=0.5)  # Optional: improves readability

plt.show()
````

## Result Explanation

The pollution category column makes the pollutant average values easier to understand. Instead of only seeing numbers, we can now see categories like Good, Moderate, Poor, Very Poor and Severe.

# 18. City-wise Average Pollution Analysis

````python
# Find average pollutant value for each city.
city_pollution = aqi_df.groupby('city')['pollutant avg'].mean().sort_values(ascending=False).head(10)

# Display top 10 cities.
city_pollution
````

````python
# Plot Top 10 Cities by Average Pollutant Value as a Pie Chart

plt.figure(figsize=(10,8))

# Function to display actual pollutant values on the pie chart
def show_values(pct):
    total = city_pollution.sum()
    value = pct * total / 100
    return f'{value:.1f}'

# Bright and dark colors
colors = plt.cm.tab10.colors

plt.pie(
    city_pollution,
    labels=city_pollution.index,
    colors=colors,
    autopct=show_values,
    startangle=90,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1},
    textprops={'fontsize': 10}
)

plt.title('Top 10 Cities by Average Pollutant Value')
plt.axis('equal')

plt.show()
````

## Result Explanation

This analysis identifies cities with the highest average pollutant readings. In this dataset, **Angul** has the highest average pollutant value among cities. This means it needs attention in pollution monitoring and dashboard analysis.

# 30. Violin Plot : Pollutant Id vs Wind Speed in Kmh

````python
# Violin plot of Wind Speed by Pollutant Type

plt.figure(figsize=(10,5))

parts = plt.violinplot(
    [aqi_weather_df[aqi_weather_df['pollutant id'] == p]['wind speed in kmh']
     for p in aqi_weather_df['pollutant id'].unique()],
    showmeans=False,
    showmedians=True
)

# Set violin colors
for body in parts['bodies']:
    body.set_facecolor('deeppink')
    body.set_edgecolor('black')
    body.set_alpha(1)

parts['cmedians'].set_color('black')
parts['cbars'].set_color('black')
parts['cmins'].set_color('black')
parts['cmaxes'].set_color('black')

plt.xticks(
    range(1, len(aqi_weather_df['pollutant id'].unique()) + 1),
    aqi_weather_df['pollutant id'].unique(),
    rotation=45
)

plt.title('Wind Speed by Pollutant Type')
plt.xlabel('Pollutant ID')
plt.ylabel('Wind Speed (km/h)')

plt.show()
````

## Result Explanation
The violin plot shows that wind speed follows a very similar pattern for all pollutant types. Most wind speed values are concentrated between 20 km/h and 27 km/h, and the median wind speed is around 21 km/h for every pollutant. This suggests that wind speed does not differ significantly among the different pollutant types in the dataset.
